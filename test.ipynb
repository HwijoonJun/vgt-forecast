```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
```

```python
model = AutoModelForSequenceClassification.from_pretrained("models/VGT_predict_finBert", num_labels=2, ignore_mismatched_sizes=True)
tokenizer = AutoTokenizer.from_pretrained("models/VGT_predict_finBert")
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
classifier("The stock price of VGT is expected to increase by 5% in the next quarter.")
```

```python
classifier("The stock price of nvidia is expected to fall by 5% in the next quarter.")
```
